### Write a function that allows you to draw a scatterplot for any given year.

```python
def scatter_plot_each_year(year):
    life_exp = df_life_exp[year]
    total_fertility_rate = df_total_fertility_rate[year]
    cols = {'life_expectancy': life_exp, 'total_fertility_rate': total_fertility_rate}
    data = pd.concat(cols, axis=1)
    data.dropna(axis=0, inplace=True)
    data.plot(kind='scatter', x='life_expectancy', 
              y='total_fertility_rate').get_figure().savefig(f"splot_{year}.png", dpi=300, bbox_inches='tight')
    plt.title(f'Plot of the fertility rate vs life expectancy in {year}')
    plt.xlabel('Life Expectancy')
    plt.ylabel('Total Fertility Rate')
    plt.plot()
```

```python
# Test
scatter_plot_each_year('1960')
```

```python
[scatter_plot_each_year(str(i)) for i in range(1960, 2011)]
```

### Convert population file function to a single value

```python
df_population.head()
```

```python
def population_convert(x):
    if x.endswith('B'):
        return float(x[:-1]) * 1_000_000_000
    elif x.endswith('M'):
        return float(x[:-1]) * 1_000_000
    elif x.endswith('k'):
        return float(x[:-1]) * 1_000
    else:
        return float(x)
    
#df_population.apply(population_convert)
df_population = df_population.applymap(population_convert)
```

```python
df_population.head()
```
